```python
ax=sb.countplot(x='Age Group', hue= "Gender", data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
sales_state= df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)

sb.set(rc={'figure.figsize' : (25,5)})  #to avoid the overlapping of location label

sb.barplot(x= 'State', y= 'Orders', data= sales_state)

```

```python
sales_amount= df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sb.set(rc={'figure.figsize' : (25,5)})  #to avoid the overlapping of location label

ax=sb.barplot(x= 'State', y= 'Amount', data= sales_amount)


```

```python
#from the above graphs we can say that from up,maharastra and karnata from these 3states most of the orders have been made and also these 3tates purchasing power is also more

Marital status
```

```python
ax=sb.countplot(x='Marital_Status', data=df)

sb.set(rc={'figure.figsize' : (8,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sb.countplot(x='Occupation', data=df)

sb.set(rc={'figure.figsize' : (15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu= df.groupby(['Occupation','Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sb.set(rc={'figure.figsize' : (20,5)})  #to avoid the overlapping of location label

ax=sb.barplot(x= 'Occupation', y= 'Amount',hue = 'Gender', data= sales_occu)
```
